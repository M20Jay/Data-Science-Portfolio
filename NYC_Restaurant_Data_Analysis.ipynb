NYC_ Restaurant Data Analysis

```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/Python Dataset/Datasets/NYC_restaurant .csv"
```

```python
df = pd.read_csv(path)
```

```python
# prompt: bar charts

import matplotlib.pyplot as plt
# Bar Charts

# Bar chart for Cuisine
plt.figure(figsize=(10, 7))
sns.barplot(x='CUISINE DESCRIPTION', y='SCORE', data=df)
plt.title('Type of violation in Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()

# Bar chart for Borough
plt.figure(figsize=(10, 7))
sns.barplot(x='Borough', y='Count', data=df)
plt.title('Number of Restaurants by Borough')
plt.xlabel('Borough')
plt.ylabel('Count')
plt.show()
```

```python
# Most common cuisines

import matplotlib.pyplot as plt
# Bar chart of the most common cuisines
cuisine_counts = df['CUISINE DESCRIPTION'].value_counts()
cuisine_counts.plot(kind='bar')
plt.xlabel('CUISINE DESCRIPTION')
plt.ylabel('Count')
plt.title('Most Common Cuisines')
plt.show()

# Bar chart of the average rating by borough
average_ratings = df.groupby('borough')['rating'].mean()
average_ratings.plot(kind='bar')
plt.xlabel('Borough')
plt.ylabel('Average Rating')
plt.title('Average Rating by Borough')
plt.show()
```

```python
# prompt: histogram

import matplotlib.pyplot as plt
# Histogram of the restaurant scores
df['SCORE'].plot.hist(bins=10)
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Restaurant Scores')
plt.show()
```

```python
# Histogram of the number of violations
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
sns.histplot(df['VIOLATION CODE'], bins=10)
plt.xlabel('Number of Violations')
plt.ylabel('Frequency')
plt.title('Distribution of Number of Violations')
plt.show()
```

```python
# Scatter plot of the score vs. the number of violations
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
sns.scatterplot(x='SCORE', y='VIOLATION CODE', data=df)
plt.xlabel('Score')
plt.ylabel('Number of Violations')
plt.title('Scatter Plot of Score vs. Number of Violations')
plt.show()
```
